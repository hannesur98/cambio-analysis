```python
import matplotlib.pyplot as plt
import hpotk
import os
hpotk.util.setup_logging()
```

```python
hpo_dir = "../../../data/hpo" ## ANPASSEN
hp_json = os.path.join(hpo_dir, "hp.json")
if not os.path.isfile(hp_json):
    raise FileNotFoundError()
fpath_ic_mica = os.path.join(hpo_dir,"term-pair-similarity.2023-10-09.csv.gz")
```

```python
from stratiphy.ingest.phenopackets import read_phenopacket_dir
path_to_cohort = 'phenopackets'
cohort = read_phenopacket_dir(path_to_cohort)
```

```python
hpo = hpotk.load_minimal_ontology(hp_json)
```

```python
from stratiphy.visualize import plot_phenotypic_feature_counts
fig, ax = plt.subplots()
plot_phenotypic_feature_counts(cohort, ax)
```
